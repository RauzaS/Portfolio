```python
# Кастомное разбиение на группы баланса
def est_salary_groupped(row):
    out = row.copy()
    out[out<1500000] = row[row<1500000] //100000 * 100000
    out[out >= 1500000] = 1500000 // 100000 * 100000
    return out
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.pointplot(data=df.assign(gr=lambda row: est_salary_groupped(row['est_salary'])),
             x='gr', y='churn', ax=ax)
ax.axhline(df['churn'].mean(), label='Total churn', ls='--', c='r')
ax.grid(c='grey', lw=0.25)
ax.legend()
plt.title('Отток по оценочной зарплате')
plt.show()
```

В интервале з/п от 100 тыс до 220 тыс есть увеличение  оттока,

## Проверка гипотез

### Гипотеза 1 : проверим различие возраста клиентов, пользующихся 2 продуктами и 1 продуктом банка

 - Н0 - возраст пользователей, пользующихся одним или двумя продуктами, одинаковый
 - Н1 - возраст пользователей , пользующихся одним или двумя продуктами, не одинаковый

```python
df['products'].value_counts()
```

Количество пользователей одним продуктом = 3306
Количество пользователей двумя продуктами = 5095

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.distplot(df[df['products'] == 1]['age'], label = 'один продукт')
sns.distplot(df[df['products'] == 2]['age'], label = 'два продукта')
plt.ylabel('количество пользователей')
plt.title ('Возраст клиентов от количества используемых продуктов банка')
plt.show()
```

Для проверки гипотезы о сравнении двух средних использую ttest. На входе имеем два списка, разной численности - клиенты, пользующиеся 1 продуктом -3306 и двумя продуктами -5095 человек. Нам нужно понять, насколько средние арифметические этих двух групп одинаковые. 

Исходные данные должны быть нормально распределены, что мы видим на графике.t-test позволяет оценить средние арифметические двух выборок (попарное сравнение) 

```python
prod_1 = df[df['products']==1]['age']
prod_2 = df[df['products']==2]['age']

alpha = 0.05
results = st.stats.ttest_ind(prod_1, prod_2, equal_var=False)

print('р-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвегаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Гипотеза о равенстве возраста при использовании одного продукта или двух продуктов банка - не подтвердилась.

### Гипотеза 2: проверим равенство долей оттока в городах Ярославль и Рыбинск

 - Н0 - доля оттока пользователей в городах Ярославль и Рыбинск одинаковые
 - Н1 - доля оттока пользователей в городах Ярославль и Рыбинск не одинаковые

Проведу z-test (проверка гипотезы о равенстве долей). Использую данные о количестве пользователей в городах и количестве пользователей, ушедших из банка. 

количество пользователей в Ярославле

```python
counts_0 = df['Yaroslavl'].sum()
counts_0
```

отток в Ярославле

```python
churn_0 = df[df['city'] =='Ярославль']['churn'].sum()
churn_0
```

количество пользователей в Рыбинске

```python
counts_1 = df['Rubinsk'].sum()
counts_1
```

отток в Рыбинске

```python
churn_1 = df[df['city'] =='Рыбинск']['churn'].sum()
churn_1
```

```python
alpha = 0.05 

p1=churn_0/ counts_0 
p2=churn_1/ counts_1
p_combined=(churn_0+churn_1) / (counts_0+counts_1)
difference = p1-p2
z_value =difference / mth.sqrt(p_combined * (1 - p_combined) * (1/counts_0 + 1/counts_1)) 

distr=st.norm(0,1) 

p_value = (1 - distr.cdf(abs(z_value))) * 2  
print('p-значение: ', p_value)

if p_value < alpha: 
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
```

z-test позволяет определить среднее значение распределения двух выборок, на входе выборка клиентов разных городов  ( Ярославля и Рыбинска) и список оттока по этим же городам. z-test может определить насколько пропорции оттока по этим двум городам статистически равные.

Гипотеза о равенстве долей оттока в городах Ярославль и Рыбинск не подтвердилась
